## Correções
Card: Corrigir as inconsistências nos dados

```python
#Has balnk
main_data.Churn = main_data['Churn'].replace([''], 'Unkonow') 

#Use 1/0 insted of Yes/No and 'No X Service' insted of 'no'
columns_to_ajust = ['customer.SeniorCitizen', 'phone.MultipleLines', 'internet.OnlineSecurity', 'internet.OnlineBackup', 'internet.DeviceProtection', 'internet.TechSupport', 'internet.StreamingTV', 'internet.StreamingMovies']
for column in columns_to_ajust:
  main_data[column] = main_data[column].replace(['No internet service', 'No phone service', 0], 'No').replace([1], 'Yes')

#Formated as String not Number
main_data['account.Charges.Total'] = main_data['account.Charges.Total'].replace([' '], '0')
main_data['account.Charges.Total'] = pd.to_numeric(main_data['account.Charges.Total'])

main_data
```

## Extra 2: Adicionar Contas Diárias
